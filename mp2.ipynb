```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt

data = {
    "Review": [
        "This product is amazing",
        "Very bad experience",
        "I love this item",
        "Not good at all",
        "Average quality",
        "Excellent service",
        "Worst product ever"
    ]
}

df = pd.DataFrame(data)

def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity
    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"

df['Sentiment'] = df['Review'].apply(get_sentiment)
print(df)

counts = df['Sentiment'].value_counts()
counts.plot(kind='bar', title='Sentiment Analysis')
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```
